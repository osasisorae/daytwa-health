```python
from selenium.webdriver.common.by import By
```

```python
def get_article_links(d):
    # Perform web automation tasks
    elements = d.find_elements(By.CLASS_NAME, a_tag)
    
    # Extract and print the href attributes of the <a> elements
    for element in elements:
        href = element.get_attribute('href')
        
        print(href)
```

```python
def summarize_article(url):

    # Get the article text using BeautifulSoup
    response = requests.get(url)
    html_content = response.content
    soup = BeautifulSoup(html_content, "html.parser")
    title = soup.title.string
    print(title)
    text = soup.get_text()

    # Generate a summary using OpenAI GPT-3.5-Turbo
    word_count = len(text.split())
    summary_length = int(word_count * 0.25)  # 25% of the word count
    prompt = f"Summarize the following article in {summary_length} words:\n{text}"
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are called daytwa. You are a helpful assistant bot built to help desigh a preventive health software."},
            {"role": "user", "content": prompt},
        ]
    )
    summary = response['choices'][0]['message']['content']

    print(summary)
    return None
```

```python
summarize_article("https://www.healthline.com/health/what-is-preventive-health-and-why-is-it-important")
```

```python
from selenium import webdriver

# Create a new instance of the Chrome driver
driver = webdriver.Chrome()

# Navigate to a website
driver.get(url)


current_page = 1

while True:
    try:
        print(f'Current Page: {current_page}'

        # Find the <ul> element with class "ais-Pagination-list"
        ul_element = driver.find_element(By.CLASS_NAME, "ais-Pagination-list")

        # Find the <a> element with the text matching the current page number
        next_page_a = ul_element.find_element(By.PARTIAL_LINK_TEXT, str(current_page))
        
        # get_article_links(driver)
       
        next_page_a.click()
        
        # Wait briefly for the page to load (you may need to adjust the wait time)
        time.sleep(2)

        # Increment the current page counter
        current_page += 1
    except Exception as e:
        # If there are no more pages, break out of the loop
        print(e)
        break

# Close the browser when done
driver.quit()
```
